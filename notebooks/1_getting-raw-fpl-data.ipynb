```python
client = APIClient()
```

### `bootstrap-static`
This is the main endpoint, which returns a nested JSON object containing the data for players, positions and teams.

```python
# get raw bootstrap static data
data = client.get_bootstrap_static()
pprint(data.model_dump(), indent=2, depth=1)
```
